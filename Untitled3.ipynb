Using the IQR, the outlier data points are the ones falling below Q1–1.5 IQR or above Q3 + 1.5 IQR. The Q1 is the 25th percentile and Q3 is the 75th percentile of the dataset, and IQR represents the interquartile range calculated by Q3 minus Q1 (Q3–Q1).[Inspired by](https://careerfoundry.com/en/blog/data-analytics/how-to-find-outliers/#:~:text=A%20box%20plot%20allows%20us,makes%20them%20easy%20to%20see)

```python
#creating a function to find the outliersin the dataframe
def find_outliers_IQR(new_data):

   q1=new_data.quantile(0.25)
   q3=new_data.quantile(0.75)
   IQR=q3-q1
   outliers = new_data[((new_data<(q1-1.5*IQR)) | (new_data>(q3+1.5*IQR)))]
   return outliers
```

```python
#Finding the outliers in the mileage column and price column
outliers_mileage = find_outliers_IQR(new_data['mileage'])
outliers_price = find_outliers_IQR(new_data['price'])
```

##Feature Engineering, Data Transformations

Replacing outliers in the dataset and Data Transfromation

```python
#replacing outliers in the mileage and price columns
def impute_outliers_IQR(new_data):
   q1=new_data.quantile(0.25)
   q3=new_data.quantile(0.75)
   IQR=q3-q1
   upper = new_data[~(new_data>(q3+1.5*IQR))].max()
   lower = new_data[~(new_data<(q1-1.5*IQR))].min()
   new_data = np.where(new_data > upper,
       new_data.median(),
       np.where(
           new_data < lower,
           new_data.median(),
           new_data
           )
       )
   return new_data
```

```python
new_data['mileage'] = impute_outliers_IQR(new_data['mileage']);
```

```python
new_data['price'] = impute_outliers_IQR(new_data['price']);
```
